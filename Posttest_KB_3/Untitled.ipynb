# Import Package Pandas, Matplotlib dan Seaborn

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

# Membaca Dataset

```python
df = pd.read_csv("netflix_titles.csv")
df
```

```python
df.dtypes
```

# Pie Chart

```python
df['rating'].value_counts()
```

```python
labels = df['rating'].value_counts().rename_axis('Rating').reset_index(name='Jumlah')
labels
```

```python
plt.figure(figsize=(8,8), dpi=300)
plt.title(' Netflix Movie & TV Show Titles ')
plt.pie(df['rating'].value_counts(), labels=labels['Rating'], autopct='%.1f%%', startangle=100)
plt.show()
```

# Grafik Boxplot

```python
sns.boxplot(x= df['release_year'], y=df['rating'])
plt.show()
```

# Grafik Scatter Plot

```python
sns.scatterplot(data = df, x='rating', y = 'release_year')
plt.title("Scatter Plot")
plt.show()
```
